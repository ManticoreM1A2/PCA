```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IEC_2017.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
result1a = df.groupby('rate_category')['consumption_unit'].aggregate(['sum'])
result1b = df.groupby('rate_category')['consumption'].aggregate(['sum'])

result1a.head()
fig1, axs = plt.subplots(2, figsize=(25,13))
fig1.suptitle('Burgers i consumed :D')

axs[0].plot(result1a, 'tab:red')
axs[0].set_title('Chicken')
axs[1].plot(result1b, 'tab:blue')
axs[1].set_title('Beef')
```

```python
result2a = df.groupby('community')['consumption_unit'].aggregate(['sum'])
result2b = df.groupby('community')['consumption'].aggregate(['sum'])

result2a.head()
fig1, axs = plt.subplots(2, figsize=(25,13))
fig1.suptitle('Burgers i consumed :D')

axs[0].plot(result2a, 'tab:red')
axs[0].set_title('Chicken')
axs[1].plot(result2b, 'tab:blue')
axs[1].set_title('Beef')
```

```python
df['consumption_unit'].value_counts().describe()
```

```python
df['consumption'].value_counts().describe()
```

```python
fig1, ax = plt.subplots(figsize=(25,7))
fig1.suptitle('Communities & Category usage unimputed')
df.groupby(['community','rate_category']).sum()['consumption_unit'].unstack().plot(ax=ax)
```

```python
fig2, ax = plt.subplots(figsize=(25,7))
fig2.suptitle('Communities & Category usage imputed')
df.groupby(['community','rate_category']).sum()['consumption'].unstack().plot(ax=ax)
```

```python
fig3, ax = plt.subplots(figsize=(25,7))
fig3.suptitle('Communities & Category usage imputed')
df.groupby(['rate_category','community']).sum()['consumption'].unstack().plot.bar(ax=ax, legend=None, width=3)
```

```python
fig4, ax = plt.subplots(figsize=(25,7))
fig4.suptitle('Communities & Category usage imputed')
df.groupby(['rate_category','community']).sum()['consumption_unit'].unstack().plot.bar(ax=ax, legend=None, width=3)
```

```python
RateCategory_2017 = df.groupby("rate_category")[["consumption_unit"]].sum().sort_values('consumption_unit',ascending=False).plot.bar()
print(RateCategory_2017)
```

```python
RateCategory_2017 = df.groupby("rate_category")[["consumption"]].sum().sort_values('consumption',ascending=False).plot.bar()
print(RateCategory_2017)
```

```python
import scipy.stats as stats
```

```python
df_mean = np.mean(df["consumption"])
df_std = np.std(df["consumption"])

pdf = stats.norm.pdf(df["consumption"].sort_values(), df_mean, df_std)

plt.plot(df["consumption"].sort_values(), pdf)
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()
```
