```python
df = pd.read_csv("IEC_2017-2022v1.csv")
```

```python
df['year'] = pd.DatetimeIndex(df['consumption_periods']).year
df['month'] = pd.DatetimeIndex(df['consumption_periods']).month
df['day'] = pd.DatetimeIndex(df['consumption_periods']).day
```

```python
RateCategory = df.groupby("community")[["consumption"]].sum().sort_values('consumption',ascending=False).plot.bar(figsize=(40,7), width=0.8, title='Communities usage all time')
print(RateCategory)
```

```python
result7a = df[df['community'] == 345].groupby(df['year'])[['consumption']].aggregate(['sum'])
result7b = df[df['community'] == 345].groupby(df['year'])[['consumption']].aggregate(['sum'])
result7c = df[df['community'] == 345].groupby(df['year'])[['consumption']].aggregate(['sum'])
result7d = df[df['community'] == 345].groupby(df['year'])[['consumption']].aggregate(['sum'])

fig3, axs = plt.subplots(4, figsize=(25,10))
fig3.suptitle('Communitie 392 usage over years')

axs[0].plot(result7a, 'tab:red')
axs[0].set_title('2017')

axs[1].plot(result7b, 'tab:green')
axs[1].set_title('2018')

axs[2].plot(result7c, 'tab:blue')
axs[2].set_title('2019')

axs[3].plot(result7d, 'tab:purple')
axs[3].set_title('2020')
```
