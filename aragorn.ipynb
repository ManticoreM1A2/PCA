```python
import pandas as pd

# https://www.dubaipulse.gov.ae/data/allresource/dewa-consumption/dewa_electricity_consumption-open?count=187&result_per_page=100&page=1
```

```python
df = pd.read_csv("EC_2017-2020.csv")
```

```python
df.sort_values('calendar_month', ascending=True)
```

```python
resul1 = df[['consumption_unit']].aggregate('sum')
print(resul1)

# Total power consumed
```

```python
result2 = df.groupby('calendar_month')['consumption_unit'].aggregate('sum')
print(result2)

# Months againest total used in said month
```

```python
result3 = df.groupby('rate_category')['consumption_unit'].aggregate('sum')
print(result3)

# Consumer categories againest total power used by said categories
```

```python
result4 = df.groupby('rate_category')['consumption_unit'].aggregate(['max', 'median', 'min', 'std', 'var'])
print(result4)

# Consumer categories with their max, min etc...
```

```python
result5 = df.groupby("community")["consumption_unit"].aggregate(['sum', 'max', 'min', 'median', 'std'])
print(result5)

# Communities with their max, min etc...
```
